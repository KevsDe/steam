```python
import pandas as pd
import numpy as np
import re
import src.cleaning as cl
```

```python
steam= pd.read_csv("Input/steam_app_data.csv",engine='python')
pd.set_option('display.max_column', 100)
```

```python
steam.head()
```

#### Analizo las columnas para decidir cuales son de utilidad antes de comenzar a trabajar en el data set.

```python
steam=steam[['platforms', 'categories', 'genres','recommendations']]
```

#### Eliminamos las filas donde faltan todos los elementos. 

```python
steam=steam.dropna(how='all')
```

```python
steam.head()
```

```python
df=steam
```

#### Extraemos de la columna platforms para que sistema operativo está disponible cada juego.

```python
df["platforms"] = df["platforms"].apply(lambda x : dict(eval(x)) )
df2 = df["platforms"].apply(pd.Series )
steam_v1 = pd.concat([df, df2], axis=1).drop('platforms', axis=1)
```

```python
dt=steam_v1
```

#### De la columna recommendations extraemos únicamente el número de recomendaciones

```python
dt['recommendations']=dt['recommendations'].apply(cl.cast)
```

```python
dt['recommendations']=dt['recommendations'].apply(cl.find_numbers)
```

```python
dt=dt
```

#### De la columna categories filtramos si es single or multi-player

```python
dt['categories']=dt['categories'].apply(cl.cast)
```

```python
dt['categories']=dt['categories'].apply(cl.cleaning)
```

#### Muchos juegos tienen varios generos, por lo que nos quedamos con los generos principales aplicando la fucnión clean.


```python
dt['genres']=dt['genres'].apply(cl.cast)
```

```python
dt['genres']=dt['genres'].apply(cl.clean)
```

```python
dt.head()
```

#### Completamos los valores NaN en categoriess, recommendations y dlc

```python
dt.categories = df.categories.fillna('Unknown')
```

```python
dt.recommendations=dt.recommendations.replace(np.nan, 0)
```

#### Renombramos las columnas

```python
dt=dt.rename(columns={"categories": "player_type", "genres": "main_genre"})
```

```python
dt.head()
```

#### Exportamos el nuevo csv

```python
dt.to_csv('steam_clean.csv', index=False)
```
